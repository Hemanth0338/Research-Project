```python
import zipfile
import os

def extract_zip_files(directory="."):
  """
  Extracts all zip files in the specified directory.

  Args:
    directory: The directory to search for zip files. Defaults to the current directory.
  """
  for filename in os.listdir(directory):
    if filename.endswith(".zip"):
      filepath = os.path.join(directory, filename)
      try:
        with zipfile.ZipFile(filepath, 'r') as zip_ref:
          zip_ref.extractall(directory)
          print(f"Extracted: {filename}")
      except zipfile.BadZipFile:
          print(f"Error: {filename} is not a valid zip file.")
      except Exception as e:
          print(f"Error extracting {filename}: {e}")

# Call the function to extract all zip files in the current directory
extract_zip_files()
```

```python
import pandas as pd
import numpy as np
from textblob import TextBlob
import spacy
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from collections import Counter

def clean_text(text):
    """
    Clean and preprocess the text data
    """
    # Convert to lowercase
    text = text.lower()

    # Remove special characters but keep important medical punctuation
    text = re.sub(r'[^a-zA-Z\s\.\,]', ' ', text)

    # Remove extra whitespace
    text = ' '.join(text.split())

    return text

def create_sentiment_pipeline(df, text_column='FOI_TEXT'):
    """
    Complete NLP pipeline for medical device incident sentiment analysis
    """
    # Initialize sentiment analyzers
    vader = SentimentIntensityAnalyzer()
    nlp = spacy.load('en_core_web_sm')

    # Results dictionary
    results = {
        'original_text': [],
        'cleaned_text': [],
        'vader_sentiment': [],
        'textblob_sentiment': [],
        'key_entities': [],
        'severity_score': [],
        'patient_impact': []
    }

    # Severity keywords
    severity_terms = {
        'high': ['died', 'expired', 'death', 'fatal', 'cardiac arrest', 'failure'],
        'medium': ['malfunction', 'error', 'fault', 'failed', 'difficulty'],
        'low': ['routine', 'maintenance', 'inspection', 'testing']
    }

    for text in df[text_column]:
        # Clean text
        cleaned_text = clean_text(text)

        # VADER sentiment
        vader_scores = vader.polarity_scores(cleaned_text)

        # TextBlob sentiment
        blob = TextBlob(cleaned_text)

        # SpaCy processing
        doc = nlp(cleaned_text)

        # Extract key medical entities
        entities = [(ent.text, ent.label_) for ent in doc.ents]

        # Calculate severity score
        severity_score = 0
        text_lower = cleaned_text.lower()

        for level, terms in severity_terms.items():
            if any(term in text_lower for term in terms):
                if level == 'high':
                    severity_score = 3
                elif level == 'medium':
                    severity_score = 2
                else:
                    severity_score = 1
                break

        # Determine patient impact
        patient_impact = 'No Impact'
        if 'no patient involvement' in text_lower:
            patient_impact = 'No Patient Involvement'
        elif 'expired' in text_lower or 'died' in text_lower:
            patient_impact = 'Fatal'
        elif 'adverse effect' in text_lower:
            if 'no adverse effect' in text_lower:
                patient_impact = 'No Adverse Effect'
            else:
                patient_impact = 'Adverse Effect'

        # Store results
        results['original_text'].append(text)
        results['cleaned_text'].append(cleaned_text)
        results['vader_sentiment'].append(vader_scores['compound'])
        results['textblob_sentiment'].append(blob.sentiment.polarity)
        results['key_entities'].append(entities)
        results['severity_score'].append(severity_score)
        results['patient_impact'].append(patient_impact)

    # Convert results to DataFrame
    results_df = pd.DataFrame(results)

    # Add summary statistics
    summary_stats = {
        'avg_vader_sentiment': results_df['vader_sentiment'].mean(),
        'avg_textblob_sentiment': results_df['textblob_sentiment'].mean(),
        'severity_distribution': results_df['severity_score'].value_counts().to_dict(),
        'patient_impact_distribution': results_df['patient_impact'].value_counts().to_dict()
    }

    return results_df, summary_stats

def analyze_sentiment_patterns(results_df):
    """
    Analyze patterns in sentiment results
    """
    # Calculate correlations
    correlations = results_df[['vader_sentiment', 'textblob_sentiment', 'severity_score']].corr()

    # Analyze sentiment by patient impact
    sentiment_by_impact = results_df.groupby('patient_impact')[['vader_sentiment', 'textblob_sentiment']].mean()

    # Find most common entities
    all_entities = [ent for entities in results_df['key_entities'] for ent in entities]
    common_entities = Counter(all_entities).most_common(10)

    return {
        'correlations': correlations,
        'sentiment_by_impact': sentiment_by_impact,
        'common_entities': common_entities
    }

# Example usage
def main(df):
    """
    Run the complete sentiment analysis pipeline
    """
    # Run main sentiment analysis
    results_df, summary_stats = create_sentiment_pipeline(df)

    # Analyze patterns
    pattern_analysis = analyze_sentiment_patterns(results_df)

    return results_df, summary_stats, pattern_analysis
```

```python
import pandas as pd
import numpy as np
from collections import Counter
import spacy
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.util import ngrams
import nltk
nltk.download('punkt')

class AdverseEventAnalyzer:
    def __init__(self):
        self.nlp = spacy.load('en_core_web_sm')

        # Define key phrases and terms to look for
        self.medical_terms = {
            'cardiac_conditions': [
                'cardiac arrest', 'ventricular fibrillation', 'vtach',
                'tachycardia', 'bradycardia', 'arrhythmia', 'heart rhythm',
                'cardiovascular', 'chest pain', 'palpitations'
            ],
            'patient_outcomes': [
                'expired', 'death', 'died', 'fatal', 'adverse effect',
                'critical condition', 'deterioration', 'complication'
            ],
            'device_issues': [
                'analysis halted', 'defib fault', 'error', 'malfunction',
                'bridge test failed', 'no video display', 'poor pad contact',
                'did not discharge', 'not shock', 'difficult to depress',
                'charge button', 'signal lost', 'device failure'
            ]
        }

    def extract_events(self, text):
        """Extract adverse events from text"""
        text = text.lower()
        events = []

        # Check for predefined terms
        for category, terms in self.medical_terms.items():
            for term in terms:
                if term in text:
                    events.append({
                        'event': term,
                        'category': category,
                        'context': self._get_context(text, term)
                    })

        return events

    def _get_context(self, text, term, window=50):
        """Get surrounding context for an event"""
        try:
            index = text.index(term)
            start = max(0, index - window)
            end = min(len(text), index + len(term) + window)
            return text[start:end]
        except ValueError:
            return ""

    def analyze_frequency(self, df, text_column='FOI_TEXT'):
        """Analyze frequency of adverse events"""
        all_events = []
        event_details = []

        for text in df[text_column]:
            events = self.extract_events(text)
            all_events.extend([(e['event'], e['category']) for e in events])
            event_details.extend(events)

        # Count frequencies
        event_freq = Counter(all_events)

        # Create detailed summary
        event_summary = []
        for (event, category), count in event_freq.most_common():
            relevant_details = [
                detail for detail in event_details
                if detail['event'] == event
            ]

            event_summary.append({
                'event': event,
                'category': category,
                'frequency': count,
                'example_contexts': [
                    detail['context'] for detail in relevant_details[:3]
                ]
            })

        return event_summary

    def identify_top_events(self, df, text_column='FOI_TEXT', n_top=15):
        """Identify top adverse events with context"""
        # Get event frequency analysis
        event_summary = self.analyze_frequency(df, text_column)

        # Organize by category
        categorized_events = {
            'cardiac_conditions': [],
            'device_issues': [],
            'patient_outcomes': []
        }

        for event in event_summary:
            if event['category'] in categorized_events:
                categorized_events[event['category']].append(event)

        # Sort within each category by frequency
        for category in categorized_events:
            categorized_events[category] = sorted(
                categorized_events[category],
                key=lambda x: x['frequency'],
                reverse=True
            )[:5]  # Top 5 from each category

        return categorized_events

def analyze_top_events(df):
    """Main function to analyze top adverse events"""
    analyzer = AdverseEventAnalyzer()

    # Get top events by category
    top_events = analyzer.identify_top_events(df)

    # Format results for presentation
    formatted_results = {
        'summary': {
            'total_cardiac_conditions': len(top_events['cardiac_conditions']),
            'total_device_issues': len(top_events['device_issues']),
            'total_patient_outcomes': len(top_events['patient_outcomes'])
        },
        'detailed_findings': top_events,
        'recommendations': []
    }

    # Generate recommendations based on findings
    for category, events in top_events.items():
        if events:  # If there are events in this category
            top_event = events[0]
            if category == 'device_issues' and top_event['frequency'] > 1:
                formatted_results['recommendations'].append(
                    f"High frequency of '{top_event['event']}' detected - "
                    f"Consider device maintenance review"
                )
            elif category == 'cardiac_conditions' and top_event['frequency'] > 1:
                formatted_results['recommendations'].append(
                    f"Multiple instances of '{top_event['event']}' reported - "
                    f"Review emergency response protocols"
                )

    return formatted_results

# Function to print results in a readable format
def print_top_events(analysis_results):
    """Print top events in a formatted way"""
    print("=== TOP ADVERSE EVENTS ANALYSIS ===\n")

    for category, events in analysis_results['detailed_findings'].items():
        print(f"\n{category.upper().replace('_', ' ')}:")
        print("-" * 50)

        for i, event in enumerate(events, 1):
            print(f"{i}. {event['event'].upper()}")
            print(f"   Frequency: {event['frequency']}")
            print(f"   Example context: {event['example_contexts'][0][:100]}...")
            print()

    print("\nRECOMMENDATIONS:")
    print("-" * 50)
    for rec in analysis_results['recommendations']:
        print(f"- {rec}")

# Example usage
def main(df):
    """Run the complete analysis"""
    results = analyze_top_events(df)
    print_top_events(results)
    return results
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

def visualize_adverse_events(results):
    """Visualizes the top adverse events."""

    # Bar plot of event frequencies
    for category, events in results['detailed_findings'].items():
        event_names = [event['event'] for event in events]
        frequencies = [event['frequency'] for event in events]

        plt.figure(figsize=(10, 6))
        sns.barplot(x=frequencies, y=event_names)
        plt.title(f"Top {category.replace('_', ' ').title()} Events")
        plt.xlabel("Frequency")
        plt.ylabel("Event")
        plt.show()

    # Pie chart of event categories
    category_counts = {
        'Cardiac Conditions': results['summary']['total_cardiac_conditions'],
        'Device Issues': results['summary']['total_device_issues'],
        'Patient Outcomes': results['summary']['total_patient_outcomes']
    }
    plt.figure(figsize=(8, 8))
    plt.pie(category_counts.values(), labels=category_counts.keys(), autopct='%1.1f%%', startangle=90)
    plt.title("Adverse Event Category Distribution")
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()

    #Word Cloud
    all_events = []
    for category, events in results['detailed_findings'].items():
        all_events.extend([event['event'] for event in events])

    text = ' '.join(all_events)
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()

# Visualize
visualize_adverse_events(results)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

def create_adverse_event_wordcloud(df, text_column='FOI_TEXT'):
    """
    Generate a word cloud visualization for adverse events mentioned in the text.

    Parameters:
    df (pandas.DataFrame): The DataFrame containing the incident report data.
    text_column (str): The name of the column containing the text data.
    """
    # Filter the DataFrame to only include rows with relevant terms
    adverse_event_terms = r'PAIN|INFECTION|SICK|BLEEDING|EROSION|SEVERE|DEMAGE|TIGHT|HEALTH PROBLEM|ABNORMAL|ANXIETY|NEGATIVE|DIFFICULT|ACHES|PAINFUL|DIARRHEA|BOWEL OBSTRUCTION|INCONTINENCE|DIED|BOWEL PROBLEMS|ANAL|DYSPAREUNIA|PAINfUL SEXUAL INTERCOURSE|REMOVAL|COME OUT|WEAKNESS|NUMBNESS'
    adverse_event_mask = df[text_column].str.contains(adverse_event_terms, case=False)
    adverse_event_df = df[adverse_event_mask]

    # Combine all the text into a single string
    adverse_event_text = ' '.join(adverse_event_df[text_column])

    # Create the word cloud
    wordcloud = WordCloud(
        width=800,
        height=400,
        stopwords=STOPWORDS,
        background_color='white',
        colormap='viridis'
    ).generate(adverse_event_text)

    # Plot the word cloud
    plt.figure(figsize=(12, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title('Adverse Events Mentioned in the Data')
    plt.show()

create_adverse_event_wordcloud(resultant_df)
```
